```python
#Write a python program to scrape first 10 product details which include product name , price , Image URL from
#https://www.bewakoof.com/women-tshirts?ga_q=tshirts . 
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html=urlopen('https://www.bewakoof.com/women-t-shirts')
bs=BeautifulSoup(html,'html.parser')
```

```python
productImage=[]
productName=[]
productPrice=[]

productCount=len(bs.find_all('img',class_="productImgTag",partial=False))
if productCount>=10:
    productCount=10
else:
    productCount=productCount
i=0
for i in  range(productCount):
    productimagedetail=bs.find_all('img',class_="productImgTag",partial=False)
    productnamedetail=bs.find_all('div',class_="productCardDetail",partial=False)
    productpricedetail=bs.find_all('span',class_="actualPriceText",partial=False)
    productImage.append(productimagedetail[i]['src'])
    productName.append(productnamedetail[i].text.split(',')[0].split('â‚¹')[0])
    productPrice.append(productpricedetail[i].text)
    i=i+1 
df=pd.DataFrame({'Product Name':productName,'Price':productPrice,'Image URL':productImage})
df    
```
