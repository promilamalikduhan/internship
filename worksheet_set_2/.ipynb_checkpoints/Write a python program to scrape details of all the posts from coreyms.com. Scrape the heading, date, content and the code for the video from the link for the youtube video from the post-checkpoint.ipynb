```python
#Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
#and the code for the video from the link for the youtube video from the post.
```

```python
from urllib.request import urlopen  
from urllib.request import Request  
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html=urlopen('https://coreyms.com/')
bs=BeautifulSoup(html,'html.parser')
```

```python
Posttime=[]
Postvideo=[]
Posttitle=[]
Postcontent=[]
articleCount=len(bs.find_all('div',class_="entry-content",partial=False))
j=0
for i in  range(articleCount):
    TitleData=bs.find_all('a',class_="entry-title-link",partial=False)
    ContentData=bs.find_all('div',class_="entry-content",partial=False)
    EntryTimeData=bs.find_all('time',class_="entry-time",partial=False)
    YoutubeCodeData=bs.find_all('span',class_="embed-youtube",partial=False)
    Posttitle.append(TitleData[i].text)
    Postcontent.append(ContentData[i].text.strip())
    Posttime.append(EntryTimeData[i].text)
    if(TitleData[i].text.split()[0]=='Update'):
        Postvideo.append('NAN')
        j=j-1
    else:
        Postvideo.append(str(YoutubeCodeData[j].iframe))
        j=j+1   
```

```python
df=pd.DataFrame({'Heading':Posttitle,'Date':Posttime,'Content':Postcontent,'Code':Postvideo})
df
```
