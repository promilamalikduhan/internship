```python
from urllib.request import urlopen  
from urllib.request import Request  
from bs4 import BeautifulSoup
import pandas as pd
```

```python
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent} 
request=Request("https://www.dineout.co.in/delhi-restaurants/buffet-special",None,headers) #The assembled request
html = urlopen(request)
bs=BeautifulSoup(html,'html.parser')
```

```python
restName=[]
restLoc=[]
restImage=[]
restrating=[]
restcuisine=[]
restCount=len(bs.find_all('div',class_="detail-info",partial=False))
i=0
for i in  range(restCount):
    restimagedetail=bs.find_all('img',class_="no-img",partial=False)
    Restaurentloc=bs.find_all('div',class_="restnt-loc ellipsis",partial=False)
    RestaurentName=bs.find_all('a',class_="restnt-name ellipsis",partial=False)
    Restaurentrating=bs.find_all('div',class_="restnt-rating rating-4",partial=False)
    Restaurentcuisine=bs.find_all('div',class_="detail-info",partial=False)
    restImage.append(restimagedetail[i]['data-src'])
    restLoc.append(Restaurentloc[i].text)
    restName.append(RestaurentName[i].text)
    restrating.append(Restaurentrating[i].text)
    restcuisine.append(Restaurentcuisine[i].text.split('|')[1])
    i=i+1 
```

```python
df=pd.DataFrame({'Restaurant name':restName,'Cuisine':restcuisine,'Location':restLoc,'Ratings':restrating,'Image URL':restImage})
df
```
