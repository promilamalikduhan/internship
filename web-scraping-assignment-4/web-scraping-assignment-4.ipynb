```python
def tableDataText(table,headflag):  
    rows = []
    if (headflag=='header'):
        trs = table.find_all('tr')   
        headerow = rowgetDataText(trs[0], 'th')
    else :
        tbody=table.find('tbody')
        trs = tbody.find_all('tr') 
        headerow = rowgetDataText(trs[0], 'th') 
    if headerow: # if there is a header row include first
        rows.append(headerow)
        trs = trs[1:]      
    for tr in trs: # for every table row
        if rowgetDataText(tr, 'td'):
            rows.append(rowgetDataText(tr, 'td'))
    return rows
def rowgetDataText(tr, coltag='td'):  # td (data) or th (header)      
    return[td.text.strip().split("\n")[0] for td in tr.find_all(coltag)]
```

Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
html=urlopen('https://www.imdb.com/list/ls095964455/')
bs=BeautifulSoup(html,'html.parser')
```

```python
Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
i=0
name_year_data=bs.find_all('h3',{ "class" : 'lister-item-header' })
genre=bs.find_all('span',{ "class" : 'genre' })
runtime=bs.find_all('span',{ "class" : 'runtime' })
rating=bs.find_all('span',{ "class" : 'ipl-rating-star__rating' })
voting=bs.find_all('span',{ "name" : 'nv' })
for i in range(len(name_year_data)):
    Year_span.append(name_year_data[i].text.split("\r")[0].split('\n')[3])
    Name.append(name_year_data[i].text.split("\r")[0].split('\n')[2])
    Genre.append(genre[i].text)
    Run_time.append(runtime[i].text)
    Ratings.append(rating[i].text)
    Votes.append(voting[i].text)
df=pd.DataFrame({'Name':Name,'Year':Year_span,'Genre':Genre,'Run Time':Run_time,'Ratings':Ratings,'Votes':Votes})
df
```

```python
srch_button=driver.find_element_by_xpath('//button[@class="fl qsbSrch blueBtn"]')
driver.execute_script("arguments[0].click();", srch_button)
```

```python
def getData(tags,tag_name):
    try:
        data = tags.find_element(By.TAG_NAME, tag_name)
        value = data.text
    except NoSuchElementException:
        value='-'
    return value
```

```python
name = []
designation = []
locations =[]
comp_name =[]
skills=[]
h=[]
h = driver.find_elements_by_xpath('//p[@class="highlightable"]')
name_data = driver.find_elements_by_xpath('//span[@class="fl ellipsis"]')
comp_data = driver.find_elements_by_xpath('//i[@class="cmpIcn"]')
designation_data = driver.find_elements_by_xpath('//span[@class="ellipsis clr"]')
skills_data = driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
i=0
for tags in h:
    locations.append(getData(tags,'small'))
    name.append(name_data[i].text)
    comp_name.append(comp_data[i].text)
    designation.append(designation_data[i].text)
    skills.append(skills_data[i].text)
    i=i+1    
df=pd.DataFrame({'Name':name,'Designation':designation,'Location':locations,'Company':comp_name,'Skills':skills})
df    
```

Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
driver.get("https://github.com/")
time.sleep(1)
explore_link = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[1]/details/summary')
driver.execute_script("arguments[0].click();", explore_link)
time.sleep(1)
trend_link = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')
driver.execute_script("arguments[0].click();", trend_link)
time.sleep(1)
```

```python
def getData(tags,tag_name):
    if tag_name=='span':
        try:
            data = tags.find_elements(By.TAG_NAME, tag_name)
            value = data[2].text
        except IndexError:
            value='-'
    else:
        try:
            data = tags.find_elements(By.TAG_NAME, tag_name)
            value = data[1].text
        except IndexError:
            value='-' 
    return value
```

```python
title=[]
desc=[]
lang=[]
nod=[]
title_tags = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]')
desc_tags = driver.find_elements_by_xpath('//p[@class="col-9 color-fg-muted my-1 pr-4"]')
all_tags = driver.find_elements_by_xpath('//div[@class="f6 color-fg-muted mt-2"]')   
i=0
for tags in all_tags:
    lang.append(getData(tags,'span'))
    nod.append(getData(tags,'a'))
    title.append(title_tags[i].text)
    desc.append(desc_tags[i].text)
    i=i+1
df=pd.DataFrame({'Title':title,'Description':desc,'Language':lang,'Number of Contributers':nod})
left_aligned_datafamename = df.style.set_properties(**{'text-align': 'right'})
display(left_aligned_datafamename)
```
