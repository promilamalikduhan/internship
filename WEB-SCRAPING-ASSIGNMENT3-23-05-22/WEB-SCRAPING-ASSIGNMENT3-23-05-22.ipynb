# Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”

```python
url='https://www.forbes.com/billionaires/'
driver.get(url)
time.sleep(2)
name=[]
rank=[]
net=[]
age=[]
country=[]
source=[]
industry=[]
while True:
    rank_data=driver.find_elements_by_xpath("//div[@class='rank']")
    name_data=driver.find_elements_by_xpath("//div[@class='personName']//div")
    net_data=driver.find_elements_by_xpath("//div[@class='netWorth']/div")
    age_data=driver.find_elements_by_xpath("//div[@class='age']//div")
    citizenship_data=driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    source_data=driver.find_elements_by_xpath("//span[@class='source-text']")
    industry_data=driver.find_elements_by_xpath("//div[@class='category']//div")

    for i in range(len(rank_data)):
        rank.append(rank_data[i].text)
        name.append(name_data[i].text)
        net.append(net_data[i].text)
        age.append(age_data[i].text)
        country.append(citizenship_data[i].text)
        source.append(source_data[i].text)
        industry.append(industry_data[i].text)  
    try:
        next_button = driver.find_element_by_xpath("//button[@class='tp-close tp-active']")
        next_button.click()
    except NoSuchElementException:
        try:
            next_button = driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']")
            driver.execute_script("arguments[0].click();",next_button)
            time.sleep(2)
        except NoSuchElementException:
            break
Billionaires=pd.DataFrame({'Rank':rank,'Names':name,'Net_worth':net,'Age':age,'Country':country,
                       'Source':source,'Industry':industry})
Billionaires              
```

# Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video



```python
url='https://www.youtube.com/watch?v=YPohR_9v6HA'
driver.get(url)
time.sleep(5)
cmt=[]
date=[]
upvote=[]
loopbreak=False
while True:
    #print("Page Scroll")
    
    #element=driver.find_element_by_xpath("/html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[8]/div[1]/div[2]/ytd-video-primary-info-renderer/div/div/div[1]/div[1]/ytd-video-view-count-renderer/span[1]")
    #element=driver.find_element_by_xpath("//div[@class='style-scope ytd-comments-entry-point-header-renderer']")
    #link=driver.execute_script("arguments[0].scrollIntoView(true);",element)
    #timeout=3
    #WebDriverWait(driver, timeout).until(element)
    comment_data=driver.find_elements_by_xpath("//yt-formatted-string[@class='style-scope ytd-comment-renderer']")
    date_data=driver.find_elements_by_xpath("//yt-formatted-string[@class='published-time-text above-comment style-scope ytd-comment-renderer']/a")
    upvote_data=driver.find_elements_by_xpath("//span[@id='vote-count-middle']")
    j=0
    for i in range(len(date_data)):
        if(len(cmt)<=500):
            cmt.append(comment_data[j+1].text)
            date.append(date_data[i].text)
            upvote.append(upvote_data[i].text)
            j=j+2
        else:
            loopbreak=True
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    if loopbreak:
        break

Comments=pd.DataFrame({'Comments':cmt,'Date when comment posted':date,'Upvote':upvote})
Comments
```

# 1 Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
2 In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.    
    

```python
expected_dlvry=[]
return_exchange=[]
price=[]
brand_name=[]
product_name_arr=[]
availability=[]
product_url=[]
link_arr=[]
driver.get('https://www.amazon.in/')
time.sleep(2)
search_field=driver.find_element(By.ID,value="twotabsearchtextbox")
product_name = str(input("Please Enter Product Name:"))
search_field.send_keys(product_name)
driver.find_element(By.ID,value="nav-search-submit-button").click()
time.sleep(2)
pagecount=3
for x in range(pagecount):
    listing_url=driver.current_url
    try:
        product_link_data=driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    except NoSuchElementException:
        product_link_data='-'
    for links in product_link_data:
        link_arr.append(links.get_attribute("href"))
    for prod_link in link_arr:
        driver.get(prod_link)
        try:
            price_data=driver.find_element_by_xpath('//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
            final_price=price_data.text
        except:
            try:
                price_data=driver.find_element_by_xpath('//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
                final_price=price_data.text.split()[0]
            except:
                final_price=''   
        if(final_price!=''):
            return_exchange_data=driver.find_elements_by_xpath('//a[@class="a-size-small a-link-normal a-text-normal"]')                                               
            product_brand_name_data=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')

            try:
                availability_data=driver.find_element_by_id('add-to-cart-button')  
                availability_status='In Stock'
            except:
                try:
                    availability_data=driver.find_element_by_id('a-autoid-0-announce')
                    availability_status='In Stock'
                except:
                    availability_status='Out of Stock'

            name_arr=product_brand_name_data.text.split(" ",1)
            brand_name_data=name_arr[0] 
            product_name_data=name_arr[1] 
            expected_dlvry_data=driver.find_element_by_xpath('//div[@id="mir-layout-DELIVERY_BLOCK-slot-DELIVERY_MESSAGE"]')
            expected_dlvry.append(expected_dlvry_data.text)
            exchange_found=False
            for z in return_exchange_data:
                if("Days" in z.text):
                    return_exchange.append(z.text)
                    exchange_found=True
                    break
            if(exchange_found==False):
                return_exchange.append("-")
            price.append(final_price)
            brand_name.append(brand_name_data)
            product_name_arr.append(product_name_data)
            availability.append(availability_status)
            product_url.append(prod_link)
    driver.get(listing_url)
    time.sleep(2)
    previous_tags=driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    previous_tags.click()
    time.sleep(2)
datafamename=pd.DataFrame({'Brand Name':brand_name,'Name of the Product':product_name_arr,'Price':price,'Return/Exchange':return_exchange,'Expected Delivery':expected_dlvry,'Availability':availability,'Product URL':product_url})
left_aligned_datafamename = datafamename.style.set_properties(**{'text-align': 'right'})
display(left_aligned_datafamename)
```
