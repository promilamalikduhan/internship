```python
import selenium
from selenium import webdriver
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome()
```

```python
job_info={1:['Data Analyst','Banglore'],2:['Data Scientist','Banglore'],3:['Data Scientist','Delhi/NCR']}
for itemKey,itemValue in job_info.items():
    job_title=[]
    job_location=[]
    company_name=[] 
    experience_required=[]
    driver.get('https://www.naukri.com/')
    search_field_designation=driver.find_element_by_class_name('suggestor-input')
    search_field_designation.send_keys(itemValue[0])
    search_field_location=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
    search_field_location.send_keys(itemValue[1])
    search_button=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
    search_button.click()
    url=driver.current_url
    driver.get(url)
    time.sleep(2)
    if(itemKey==3):
        element=driver.find_element(By.ID,value="chk-3-6 Lakhs-ctcFilter-")
        driver.execute_script("arguments[0].click();", element) 
        time.sleep(2)
    title_tags=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
    company_tags=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
    job_tags=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
    experience_tags=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
    elementCount=len(title_tags[0:10])
    for i in range(elementCount):
        title=title_tags[i].text
        job_title.append(title)
        cname=company_tags[i].text
        company_name.append(cname)
        jloc=job_tags[i].text
        job_location.append(jloc)
        exp=experience_tags[i].text
        experience_required.append(exp) 
    datafamename=itemKey
    datafamename=pd.DataFrame({'Title':job_title,'Location':job_location,'Company Name':company_name,'Experience Required':experience_required})
    left_aligned_datafamename = datafamename.style.set_properties(**{'text-align': 'right'})
    display(left_aligned_datafamename)
#driver.close()       
```

```python
driver.get('https://www.myntra.com/shoes')
time.sleep(2)
priceFilter=driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/input")
driver.execute_script("arguments[0].click();", priceFilter) 
time.sleep(2)
colorFilter=driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/input")
driver.execute_script("arguments[0].click();", colorFilter) 
time.sleep(2)
noOfPages=2
brand=[]
shoes_desc=[]
price=[]
for i in range(noOfPages):
    flag=False
    brand_tags=driver.find_elements(By.CLASS_NAME,value='product-brand')
    price_tags=driver.find_elements(By.CLASS_NAME,value='product-price')
    product_desc_tags=driver.find_elements(By.CLASS_NAME,value='product-product')        
    tagCount=len(brand_tags)
    for i in range(tagCount):
        if(len(brand)<=99):
            brand_data=brand_tags[i].text
            brand.append(brand_data)            
            price_data=price_tags[i].text.split()[1].split("Rs.")[0]
            price.append(price_data)
            desc_data=product_desc_tags[i].text
            shoes_desc.append(desc_data) 
        else:
            flag=True 
            break
    if(flag):
        break
    driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[2]/div/div[2]/section/div[2]/ul/li[12]/a').click() 
    time.sleep(2)
datafamename=pd.DataFrame({'Brand':brand,'Product Description':shoes_desc,'Price':price})
left_aligned_datafamename = datafamename.style.set_properties(**{'text-align': 'right'})
display(left_aligned_datafamename)
```

Q9: Write a python program to scrape data for first 10 job results for Data Scientist Designation in Noida

Q10: Write a python program to scrape the salary data for Data Scientist designation.
You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary.

```python
job_title=[]
company_name=[] 
experience_required=[]
min_salary=[]
avg_salary=[]
max_salary=[]

driver.get('https://www.ambitionbox.com/')
time.sleep(5)
driver.find_element(By.XPATH,value='/html/body/div[1]/nav/nav/a[4]').click()
url=driver.current_url
driver.get(url)
time.sleep(2)
search_field_salaries=driver.find_element(By.XPATH,value="/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
search_field_salaries.send_keys('Data Scientist')
time.sleep(2)
auto_suggest=driver.find_elements(By.CLASS_NAME,value='tt_text')
auto_suggest[0].click()
time.sleep(1)

company_name_tags=driver.find_elements(By.CLASS_NAME,value='name')


title_exp_tags=driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')

#print(title_exp_tags[0][2].text)
salary_tags=driver.find_elements_by_xpath('//div[@class="salary-values"]')
avg_salary_tags=driver.find_elements_by_xpath('//p[@class="averageCtc"]')


elementCount=len(avg_salary_tags)
j=0
for i in range(elementCount):
    if(j<=9):        
        job_title_data=title_exp_tags[i].text.split("\r")[0].split("\n")[0]
        job_title.append(title_exp_tags[i].text.split("\r")[0].split("\n")[0])
        
        campany_name_data=company_name_tags[i].text.split("\n")[0]
        company_name.append(campany_name_data)
        
        exp_required_data=title_exp_tags[i].text.split("\r")[0].split("\n")[2]
       
        experience_required.append(exp_required_data)
    
        min_sal=salary_tags[i].text.split("\n")[0]
        min_salary.append(min_sal)
        avg_sal=avg_salary_tags[i].text
        avg_salary.append(avg_sal)
        max_sal=salary_tags[i].text.split("\n")[1]
        max_salary.append(max_sal)
        time.sleep(2)
    j=j+1
    
datafamename=pd.DataFrame({'Title':job_title,'Company Name':company_name,'Experience Required':experience_required,
                          'Minimum Sal':min_salary,'Avarage Sal':avg_salary,'Maximum Sal':max_salary})
left_aligned_datafamename = datafamename.style.set_properties(**{'text-align': 'right'})
display(left_aligned_datafamename)
#driver.close()       
```
