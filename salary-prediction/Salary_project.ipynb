```python
#scales the data.essentially returns the Z-scores of every attribute

#Z-score also calles a standard score.

from sklearn.preprocessing import StandardScaler
std_scale=StandardScaler()
```

```python
df['salary']=std_scale.fit_transform(df[['salary']])
df['yrs.since.phd']=std_scale.fit_transform(df[['yrs.since.phd']])
df['yrs.service']=std_scale.fit_transform(df[['yrs.service']])
#df['rank']=std_scale.fit_transform(df[['rank']])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X=df.iloc[: , : -1].values
Y=df.iloc[: , 5].values
```

```python
len(X)
```

```python
len(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.5,random_state = 0)
```

```python
regressor  = LinearRegression()
regressor.fit(X_train,Y_train)
print('Training Complete')
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

```

```python
y_pred = regressor.predict(X_test)
```

```python
cmp = pd.DataFrame({'Actual' : Y_test, 'Predicted' : y_pred})
cmp
```
